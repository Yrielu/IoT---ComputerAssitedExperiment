Defining Some Basic Functions


```python
def shockley(Vd,Is,n):
    "Shockley diode equation, used for curve fitting, assuming T = 288K, it's quite cold"
    temp_1 = n*0.0248
    temp_2 = Vd/temp_1
    temp_3 = math.e**temp_2
    temp_4 = temp_3-1
    return Is*temp_4
```

```python
def get_cov(inp):
    "A quick way of calculating Covariance from CoV Matrix. it takes in the input of a 2*2 matrix, add returns the determinant"
    return (inp[1][0][0]*inp[1][1][1]-inp[1][0][1]**2)**0.5
```

```python
    def set_vol (vol,ser):
        'Sets voltage: by sending encoded value of vol, and decoding incoming confirmation, returning the confirmation message after set voltage'
        #Sending input vol
        arg = '<S' + str(round(vol,2))+'>'
        ser.write(arg.encode())
        #Decoding received confirmation 
        a = ser.readline()
        readstring = a.decode("utf-8")
        
        return readstring
       
        
    def get_vol (pin_point,ser):
        'Meassures voltage at specific pin point, truns the measured voltage value'
        #Sending input vol
        arg = '<V' + str(pin_point)+'>'
        ser.write(arg.encode())
        #Decoding received confirmation 
        a = ser.readline()
        readstring = a.decode("utf-8")
        
        return readstring
        
        
    def get_cur (ser):  
        'Gives calculated current from: I=(A0-A1)/R where are is set to 100, returns the current value through R'
        writestring = '<I1>'
        bytestowrite = writestring.encode()
        ser.write(bytestowrite)
        a = ser.readline()
        readstring = a.decode("utf-8")
        
        return readstring
        
        
    def clean_list(listname):
        'Cleans given list by removing strings part and converting number to float. the list writing by the main func is messy, this will return a neat list'
        split_list=[]
        for element in listname:
            split_list+=element.split(' ')
        final_list=[]    
        i=0
        while i<len(split_list):
            if i%2!=0:
                final_list+=[float(split_list[i])]
            i+=1
        return final_list
```

```python
def ivgraph(filename, voltage_start, voltage_end,n_meassurements,timesleep,repeats,ser):
    'Will set start&final voltage with n_measurements in between. can adjust sleeptime or number of repeats, and stroing info in a generated file with given filename'
    'will generate a output file, one list with all the voltages across LED after being averaged, the current being averaged. also will output the std_deviation for all measured data'
    print('Order received!')
    'creating file'
    txtfile = open(filename,"w")
    txtfile.write("\n")
    
    
    'generating all voltages needed for input'
    voltage_range=linspace(voltage_start,voltage_end, n_meassurements)  #This is voltage we will put in S command    
    print("I'll now start meassuring data")
    
    
    'Data collecting loop, a double loop, outer loop controls the number of repeats, the first inner loop goes through all of the voltage needed for input'
    'Using set_vol to set voltage, and get_vol and get_cur to recieve current and voltage info within the circuit. The second inner loop generates the file for output'
    i=1
    while i<=repeats:
        set_voltage=[]
        voltage1=[]
        voltage2=[]
        current=[]
        for each_vol in voltage_range:
            set_voltage+=[set_vol(each_vol,ser)]  #This list has confirmation voltage from S command
            time.sleep(timesleep) 
            voltage1+=[get_vol(1,ser)]
            voltage2+=[get_vol(2,ser)]  #This is meassured voltage with V2 (at A1)
            current+=[get_cur(ser)]   
        clean_set_voltage=clean_list(set_voltage)
        clean_voltage1=clean_list(voltage1)
        clean_voltage2=clean_list(voltage2)
        clean_current=clean_list(current)    
        set_vol(0,ser)
        print('Meassurements of repeat '+str(i)+' done!')        
        txtfile.write("Meassurement number: "+str(i)+"\n"+"S\t V1\t V2\t I\n")    #Writing into txt file, and adding data in columns
        for index in range(len(clean_set_voltage)):
            txtfile.write(str(clean_set_voltage[index])+"\t"+str(clean_voltage1[index])+"\t"+str(clean_voltage2[index])+"\t"+str(clean_current[index])+"\n")        
        txtfile.write("\n")
        i+=1
        
        
    'a few lines indicating progress, because some measurement can take too long, and we get impatient and want to terminate the program'
    txtfile.write("Meassurements taken from set voltage "+str(voltage_start)+" to "+str(voltage_end)+"\n")
    txtfile.write(str(n_meassurements)+" meassurements where taken between those voltages\n")
    txtfile.write("Timesleep set to: "+str(timesleep)+", and repeats to: "+str(repeats)+'\n')
    txtfile.write('measurement taken on '+str((datetime.datetime.now ())))
    txtfile.close()
    
    
    
    'Getting data from txtfile, storing V2, voltage across LED and I, into two arrays'
    all_v2=[]   
    i=0
    all_curr = []
    while i<repeats:
        if i==0:
            data_v2=loadtxt(filename ,skiprows=3, max_rows=n_meassurements, usecols=2)
            data_curr=loadtxt(filename ,skiprows=3, max_rows=n_meassurements, usecols=3)
        if i!=0:
            skip=3*(1+i)+n_meassurements*i
            data_v2=loadtxt(filename,skiprows=skip, max_rows=n_meassurements, usecols=2)
            data_curr=loadtxt(filename,skiprows=skip, max_rows=n_meassurements, usecols=3) 
        all_v2+=[list(data_v2)]
        all_curr+=[list(data_curr)]
        i+=1
    
    'Processing data to be able to calculate avg and std (changing columns=rows)'
    #print(data_v2)
    v2_by_vol=[]
    curr_by_vol=[]
    i=0    
    while i<n_meassurements: 
        v2_by_vol+=[list(array(all_v2)[:,i])]
        curr_by_vol+=[list(array(all_curr)[:,i])]
        i+=1
    #Calculate and store avg/std for v2 and curr

    'Calculate and store avg/std for v2 and curr'
    v2_avg=[]
    v2_std=[]
    for each_v in array(v2_by_vol):
        v2_avg+=[mean(each_v)]
        v2_std+=[std(each_v)]
    curr_avg=[]
    curr_std=[]  
    for each_curr in array(curr_by_vol):
        curr_avg+=[mean(each_curr)]
        curr_std+=[std(each_curr)]
    
    print("I'm done! :)")
    return array(v2_avg),array(curr_avg),array(curr_std)
```
